```python
# For creating raw text dataframe

def text_to_DataFrame(raw_text):
    text_list = list(raw_text)
    
    
    anno_dict = {
        "text":text_list,
        "rater_c":"Y",
        "rater_d":"Y"}
    df = pd.DataFrame(anno_dict)

    return df
    
```

```python
# Find contents except the span words and create it into a dataframe

def outside_span(raw_data):
    result = []
    span_tag = raw_data.find_all(
   "span")
    
    for i in span_tag:
        anno_list=[]
        anno_list.append(i.text)
        for j in span_tag:
            if j.text.strip() == anno_list[0]:
                result.append(j.next_sibling)
    
    outside_content = list(result)
    
    df_2 = pd.DataFrame(outside_content, columns = ['outside'])
   
        
    return df_2
```

```python
# Find outside contents index



def outside_id(outside_span_df):
    
    sub_index = []
    for idx in outside_span_df['outside']:
        sub_index.append(text_raw.index(idx))
        
    length_list = []
    for el in outside_span_df['outside']:
        length_list.append(len(el)-1)
    
    
    out_index = [[sub_index[a], sub_index[a]+length_list[a]] for a in range(min(len(length_list),len(sub_index),len(outside_span_df)))]

    return out_index
```

```python
# For extending ID

def extend_index(list_out):
    combined = []
    for q in list_out:
        u=[]
        if len(q)==2:
            for p in range (q[0],q[1]+1):
                u.append(p)
                #print(q)
                #print(u)
        else:
            u.append(q[0])
        q=[]
        q=u
        combined.append(u)
    
    List = [item for sublist in combined for item in sublist]

    return List
```

```python
# Find index of contents that before the first span word

def toFirst_span(df_raw):
    tag = []
    span_tag = df_raw.find_all(
   "span")
    for x in span_tag:
        tag.append(x.text)
    
    sub_l = []
    for s in tag:
        sub_l.append(text_raw.index(s))
    
    sub = sub_l[0]
    
    to_first = []
    for leng in range(0, sub, 1):
        to_first.append(leng)
    
    return to_first
    
```
